```python
import pandas as pd
import duckdb
```

```python
df = pd.read_csv('../data/state_lookup.csv')
df.head()
```

```python
df_fips = pd.read_csv('./state_fips.txt', delimiter='|')
df_fips.head()
```

```python
# join the two dataframes on df['id'] == df_fips['STATE']
df_joined = df.merge(df_fips, left_on='id', right_on='STATE')

# get only columns: df['id'], df['state'], df_fips['STATEFP']
df_joined = df_joined[['id', 'STATEFP', 'state' ]]
df_joined.head(10)
```

```python
# pad the STATEFP with a leading zero
df_joined['STATEFP'] = df_joined['STATEFP'].apply(lambda x: str(x).zfill(2))
df_joined.head(10)
```

```python
df_joined.to_csv('../data/state_fips_lookup.csv', index=False)
```
