```python
import pandas as pd
import duckdb
```

```python
df = pd.read_csv('./county_fips.txt', delimiter='|')
df['STATEFP'] = df['STATEFP'].astype(str)
df['COUNTYFP'] = df['COUNTYFP'].astype(str)
df.head()
```

```python
duckdb.sql("SELECT * FROM df limit 10")
```

```python
# count of rows
print(df.shape[0])
```

```python
query = """
SELECT 
STATE as state
, lpad(STATEFP, 2, '0') as state_fips
, lpad(COUNTYFP, 3, '0') as county_fips
, concat(lpad(STATEFP, 2, '0'), lpad(COUNTYFP, 3, '0')) as fips
, COUNTYNAME as county
FROM df 
"""
county_fips_df = duckdb.query(query).df()
county_fips_df
```

```python
county_fips_df.to_csv('../data/county_fips.csv', index=False)
```
